```python
import functools

import pandas as pd
```

```python
bldg_features_filepath = "../data/interim/bldg-features.csv"
tree_features_filepath = "../data/interim/tree-features.csv"
elev_features_filepath = "../data/interim/elev-features.csv"
lake_features_filepath = "../data/interim/lake-features.csv"

station_id_col = "station_id"
dst_filepath = "../data/processed/station-features.csv"
```

```python
station_features_df = functools.reduce(
    lambda left, right: pd.merge(left, right, on=station_id_col),
    [
        pd.read_csv(features_filepath)
        for features_filepath in [
            bldg_features_filepath,
            tree_features_filepath,
            elev_features_filepath,
            lake_features_filepath,
        ]
    ],
)
station_features_df
```

```python
# dump to file
station_features_df.to_csv(dst_filepath, index=False)
```
